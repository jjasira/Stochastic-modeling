```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
tickers = ['TLT', 'GLD', 'SPY', '^VIX']

data = yf.download(tickers, start='2005-01-01', progress=False)

# Handle both single- and multi-index formats
if 'Adj Close' in data.columns:
    data = data[['Adj Close']]
elif 'Close' in data.columns:
    data = data[['Close']]
else:
    raise KeyError("Neither 'Adj Close' nor 'Close' found in downloaded data.")

# Drop rows with missing values (ensure common sample period)
data = data.dropna()
data.head()
```

```python
# Compute daily log returns for ETFs
etf_returns = np.log(data[['TLT', 'GLD', 'SPY']] / data[['TLT', 'GLD', 'SPY']].shift(1))

# Compute ΔVIX (simple difference)
vix_change = data['^VIX'].diff()

# Alternatively, compute VIX percentage return:
vix_return = np.log(data['^VIX'] / data['^VIX'].shift(1))
```
